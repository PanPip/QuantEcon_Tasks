```python
# Speeding up the next code using NumPy
import numpy as np
from numpy.random import uniform

def sample(q):
    a = 0.0
    U = uniform(0, 1)
    for i in range(len(q)):
        if a < U <= a + q[i]:
            return i
        a = a + q[i]
```

```python
class discreteRV:
    """
    A simple example sone through class
    """

    def __init__(self,q):
        """
        Initialization and creation of cumsum element
        """
        self.q = q
        self.cumq = np.cumsum(q)
        
    def draw(self,k):
        """
        Outputs k elements from the distribution
        """
        print(self.cumq)
        res = self.cumq.searchsorted(uniform(0,1,size=k))
        return res
```

```python
#Initial code

class ECDF:

    def __init__(self, observations):
        self.observations = observations

    def __call__(self, x):
        counter = 0.0
        for obs in self.observations:
            if obs <= x:
                counter += 1
        return counter / len(self.observations)

# == test == #

samples = [uniform(0, 1) for i in range(10)]
F = ECDF(samples)

print(F(0.5))  # Evaluate ecdf at x = 0.5

```

```python
#Applied changes

class ECDF:

    def __init__(self, observations):
        self.observations = np.array(observations)

    def __call__(self, x):
        return sum(self.observations <= x)/len(self.observations)
```

```python
F = ECDF(samples)

print(F(0.5))  # Evaluate ecdf at x = 0.5
```

#### Exercise 1

Write a recursive implementation of the bisection function described below.
Test it on the function discussed above

```python
#Ex1

#Original code
def bisect1(f, a, b, tol=10e-5):
    """
    Implements the bisection root finding algorithm, assuming that f is a
    real-valued function on [a, b] satisfying f(a) < 0 < f(b).
    """
    lower, upper = a, b

    while upper - lower > tol:
        middle = 0.5 * (upper + lower)
        # === if root is between lower and middle === #
        if f(middle) > 0:
            lower, upper = lower, middle
        # === if root is between middle and upper  === #
        else:              
            lower, upper = middle, upper

    return 0.5 * (upper + lower)

#Recursive implementation
def bisect(f, lower, upper, tol=10e-5):
    """
    Implements a recursive bisection root finding algorithm, 
    assuming that f is a real-valued function on [a, b] 
    satisfying f(a) < 0 < f(b).
    """
    middle = 0.5 * (upper + lower)
    if f(middle) > 0:
        lower, upper = lower, middle
    else:              
        lower, upper = middle, upper
    if upper - lower > tol:
        return bisect(f,lower,upper,tol)
    else:
        return 0.5 * (upper + lower)
```

```python
#Unifunc examples

def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

grid = np.linspace(-3, 3, 1000)
m = -np.inf

qe.tic()
for x in grid:
    for y in grid:
        z = f(x, y)
        if z > m:
            m = z

qe.toc()
```

```python
# Finally, Numba

import matplotlib.pyplot as plt

#On the example code

def qm(x0, n):
    x = np.empty(n+1)
    x[0] = x0
    for t in range(n):
        x[t+1] = 4 * x[t] * (1 - x[t])
    return x

x = qm(0.1, 250)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, 'b-', lw=2, alpha=0.8)
ax.set_xlabel('time', fontsize=16)
plt.show()
```

```python
from numba import jit

qm_numba = jit(qm)  # qm_numba is now a 'compiled' version of qm
```

```python
qe.util.tic()
qm(0.1, int(10**5))
time1 = qe.util.toc()
```

```python
qe.util.tic()
qm_numba(0.1, int(10**5))
time2 = qe.util.toc()
```

```python
#As we will see, the consecutive executions have higher speed

qe.util.tic()
qm_numba(0.1, int(10**5))
time2 = qe.util.toc()
```

```python
time1 / time2  # Calculate speed gain
```

```python
#We can use simply as a decorator

@jit
def qm(x0, n):
    x = np.empty(n+1)
    x[0] = x0
    for t in range(n):
        x[t+1] = 4 * x[t] * (1 - x[t])
    return x
```

```python
#A catch on global variables - Numba treats global variables as constants to ensure stability

#Numba can creeate custom ufuncs

from numba import vectorize

@vectorize
def f_vec(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

grid = np.linspace(-3, 3, 1000)
x, y = np.meshgrid(grid, grid)

np.max(f_vec(x, y))  # Run once to compile

qe.tic()
np.max(f_vec(x, y))
qe.toc()
```

```python
#We can compute in parallel, but need to specify the types

@vectorize('float64(float64, float64)', target='parallel')
def f_vec(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

np.max(f_vec(x, y))  # Run once to compile

qe.tic()
np.max(f_vec(x, y))
qe.toc()
```

```python
#Cython examples

#Loop implementation for sum of geometric progression in Python

def geo_prog(alpha, n):
    current = 1.0
    sum = current
    for i in range(n):
        current = current * alpha
        sum = sum + current
    return sum
```

```python
#Now, the same in Cython

%load_ext Cython
```

```python
%%cython
def geo_prog_cython(double alpha, int n):
    cdef double current = 1.0
    cdef double sum = current
    cdef int i
    for i in range(n):
        current = current * alpha
        sum = sum + current
    return sum
```

```python
import quantecon as qe
qe.util.tic()
geo_prog(0.99, int(10**6))
qe.util.toc()
```

```python
qe.util.tic()
geo_prog_cython(0.99, int(10**6))
qe.util.toc()
```

```python
#Joblib - Python library for caching and parallelization

#Using it one can save the results of computations, for example, wheh wanting to restart the Kernel

#Example of how one can cache the results to use them later
from joblib import Memory

memory = Memory(location='./joblib_cache')

@memory.cache
def qm(x0, n):
    x = np.empty(n+1)
    x[0] = x0
    for t in range(n):
        x[t+1] = 4 * x[t] * (1 - x[t])
    return np.mean(x < 0.1)
```

```python
qe.util.tic()
n = int(1e7)
qm(0.2, n)
qe.util.toc()
```

```python
#Ex1

# Let  "0"=High; "1"= Low 

#Matrix of states
state = np.array([[0.8, 0.2],[0.1, 0.9]])
#Probability of going to state 1 (from state 0 and 1)
p = np.array([0.2, 0.9])
N = 100000

def simple_markov(p, N):
    time_in_low = 0
    state = 0
    for i in range(N):
        new_state = np.random.binomial(1, p[state])
        if np.random.binomial(1, p[state]) == 1:
            time_in_low += 1
        state = new_state
    return time_in_low/N

print('Appr. time in low: ',simple_markov(p, N) )
    
    
```
