```python
#Plot correlated time series

import numpy as np
import matplotlib.pyplot as plt

N = 100 # Size of TS
Î± = 0.9 # Dependence
init = 1 # Initial value
Ser = [init]

def GenerateTS(N,alpha):
    Ser = [0]
    for i in range(1,N):
        Ser.append(Ser[i-1]*alpha + np.random.randn())
    return(Ser)
```

```python
# replacing np.poly1d

import numpy as np

def p(x,coef):
    """
    X - a np vector
    coef - a np vector of coeffs
    """
    
    return sum(np.cumprod(x) * coef)

x = np.array([1,2,3])
coef = np.array([1,0,1])

p(x,coef)
    
```

```python
# Speeding up the next code using NumPy

from random import uniform

def sample(q):
    a = 0.0
    U = uniform(0, 1)
    for i in range(len(q)):
        if a < U <= a + q[i]:
            return i
        a = a + q[i]
```
